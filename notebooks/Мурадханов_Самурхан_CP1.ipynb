```python
print("=" * 60)
print("ОЧИСТКА ДАННЫХ (ПОЛНАЯ ВЕРСИЯ)")
print("=" * 60)

initial_rows = len(df)
print(f"\n1. НАЧАЛЬНОЕ СОСТОЯНИЕ:")
print(f"   - Всего строк: {initial_rows:,}")
print(f"   - Всего колонок: {df.shape[1]}")

print(f"\n2. ДУБЛИКАТЫ:")
duplicates_count = df.duplicated().sum()
print(f"   - Полных дубликатов строк: {duplicates_count:,}")
if duplicates_count > 0:
    df = df.drop_duplicates()
    print(f"   - Удалено дубликатов: {duplicates_count:,}")

print(f"\n3. ПРОПУСКИ (ДО УДАЛЕНИЯ):")
missing_before = df.isnull().sum()
missing_before = missing_before[missing_before > 0]
if len(missing_before) > 0:
    for col, count in missing_before.items():
        print(f"   - {col}: {count:,} ({count/initial_rows*100:.2f}%)")
else:
    print(f"   - Пропусков нет")

print(f"\n4. НЕКОРРЕКТНЫЕ ЗНАЧЕНИЯ (ДО ФИЛЬТРАЦИИ):")
negative_quantity = (df['Quantity'] < 0).sum()
negative_price = (df['UnitPrice'] < 0).sum()
zero_quantity = (df['Quantity'] == 0).sum()
zero_price = (df['UnitPrice'] == 0).sum()

print(f"   - Quantity < 0 (возвраты): {negative_quantity:,}")
print(f"   - UnitPrice < 0: {negative_price:,}")
print(f"   - Quantity = 0: {zero_quantity:,}")
print(f"   - UnitPrice = 0: {zero_price:,}")

cancelled_invoices = df[df['InvoiceNo'].astype(str).str.contains('C', na=False)].shape[0]
print(f"   - Отмененные заказы (InvoiceNo с 'C'): {cancelled_invoices:,}")

print(f"\n5. ВЫБРОСЫ (ДО ФИЛЬТРАЦИИ):")
q1_quantity = df['Quantity'].quantile(0.25)
q3_quantity = df['Quantity'].quantile(0.75)
iqr_quantity = q3_quantity - q1_quantity
outliers_quantity = ((df['Quantity'] < q1_quantity - 1.5 * iqr_quantity) |
                      (df['Quantity'] > q3_quantity + 1.5 * iqr_quantity)).sum()
print(f"   - Выбросы по Quantity: {outliers_quantity:,} ({outliers_quantity/initial_rows*100:.2f}%)")

q1_price = df['UnitPrice'].quantile(0.25)
q3_price = df['UnitPrice'].quantile(0.75)
iqr_price = q3_price - q1_price
outliers_price = ((df['UnitPrice'] < q1_price - 1.5 * iqr_price) |
                   (df['UnitPrice'] > q3_price + 1.5 * iqr_price)).sum()
print(f"   - Выбросы по UnitPrice: {outliers_price:,} ({outliers_price/initial_rows*100:.2f}%)")

print(f"\n6. ПРИМЕНЕНИЕ ФИЛЬТРОВ:")
before_filter = len(df)
df = df[df['Quantity'] > 0]
after_quantity = len(df)
print(f"   - После удаления Quantity <= 0: удалено {before_filter - after_quantity:,} строк")

df = df[df['UnitPrice'] > 0]
after_price = len(df)
print(f"   - После удаления UnitPrice <= 0: удалено {after_quantity - after_price:,} строк")

print(f"\n7. ПРЕОБРАЗОВАНИЕ ДАТЫ:")
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
print(f"   - Диапазон дат: {df['InvoiceDate'].min()} до {df['InvoiceDate'].max()}")

print(f"\n8. СОЗДАНИЕ ПРИЗНАКА TransactionAmount:")
df['TransactionAmount'] = df['Quantity'] * df['UnitPrice']
print(f"   - Средняя сумма транзакции: £{df['TransactionAmount'].mean():.2f}")
print(f"   - Максимальная сумма транзакции: £{df['TransactionAmount'].max():.2f}")

print(f"\n9. ПРОПУСКИ В CustomerID (ДО УДАЛЕНИЯ):")
customerid_missing = df['CustomerID'].isnull().sum()
print(f"   - Строк без CustomerID: {customerid_missing:,} ({customerid_missing/len(df)*100:.2f}%)")

before_customer = len(df)
df_with_customers = df.dropna(subset=['CustomerID'])
after_customer = len(df_with_customers)
print(f"   - После удаления строк без CustomerID: удалено {before_customer - after_customer:,} строк")

print(f"\n10. ИТОГОВАЯ СТАТИСТИКА ОЧИСТКИ:")
total_removed = initial_rows - len(df_with_customers)
print(f"    - Всего удалено строк: {total_removed:,}")
print(f"    - Процент удаленных строк: {total_removed/initial_rows*100:.2f}%")
print(f"    - Осталось строк: {len(df_with_customers):,}")
print(f"    - Уникальных клиентов: {df_with_customers['CustomerID'].nunique():,}")

df_clean = df_with_customers.copy()
print(f"\n✅ Очищенный датасет сохранен в переменную 'df_clean'")
```

**Feature Engineering и работа с признаками**

Этот блок создает новые признаки, которые помогут модели предсказывать месячные траты клиентов. Исходный датасет имел 8 колонок. После агрегации и создания признаков количество увеличивается до 20+. Создаются лаговые признаки

```python
print("=" * 60)
print("БЛОК 5 (ИСПРАВЛЕННЫЙ): FEATURE ENGINEERING БЕЗ DATA LEAKAGE")
print("=" * 60)

print("\n1. ИСХОДНЫЕ ПРИЗНАКИ В ДАТАСЕТЕ:")
print(f"   - Количество исходных колонок: {df_clean.shape[1]}")

print("\n2. АГРЕГАЦИЯ ПО КЛИЕНТАМ И МЕСЯЦАМ:")
df_clean['InvoiceYearMonth'] = df_clean['InvoiceDate'].dt.to_period('M')

customer_monthly = df_clean.groupby(['CustomerID', 'InvoiceYearMonth']).agg({
    'TransactionAmount': 'sum',
    'InvoiceNo': 'nunique',
    'Quantity': 'sum',
    'StockCode': 'nunique'
}).reset_index()

customer_monthly.columns = ['CustomerID', 'Month', 'TotalSpent', 'OrdersCount', 'TotalQuantity', 'UniqueProducts']
print(f"   - После агрегации: {customer_monthly.shape[0]:,} строк")

print("\n3. СОРТИРОВКА ПО ВРЕМЕНИ:")
customer_monthly = customer_monthly.sort_values(['CustomerID', 'Month'])

print("\n4. ЛАГОВЫЕ ПРИЗНАКИ (только прошлые данные):")
for lag in [1, 2, 3]:
    customer_monthly[f'Spent_Lag_{lag}'] = customer_monthly.groupby('CustomerID')['TotalSpent'].shift(lag)
    print(f"   - Spent_Lag_{lag}: траты {lag} месяц(ев) назад")

customer_monthly['Spent_MA_3'] = customer_monthly.groupby('CustomerID')['TotalSpent'].transform(
    lambda x: x.shift(1).rolling(3, min_periods=1).mean()
)
print(f"   - Spent_MA_3: скользящее среднее за 3 предыдущих месяца")

customer_monthly['Spent_Trend'] = customer_monthly['Spent_Lag_1'] - customer_monthly['Spent_Lag_2']
print(f"   - Spent_Trend: тренд трат")

print("\n5. КУМУЛЯТИВНЫЕ ПРИЗНАКИ (накопленная история ДО текущего месяца):")

cumulative_spent = df_clean.groupby('CustomerID').apply(
    lambda x: x.sort_values('InvoiceDate')['TransactionAmount'].shift(1).cumsum()
).reset_index()
cumulative_spent.columns = ['CustomerID', 'level_1', 'CumulativeSpent']
cumulative_spent = cumulative_spent.drop(columns=['level_1'])

cumulative_orders = df_clean.groupby('CustomerID').apply(
    lambda x: x.sort_values('InvoiceDate')['InvoiceNo'].shift(1).expanding().apply(
        lambda y: len(set(y)), raw=False
    ).fillna(0)
).reset_index()
cumulative_orders.columns = ['CustomerID', 'level_1', 'CumulativeOrders']
cumulative_orders = cumulative_orders.drop(columns=['level_1'])

df_clean['CumulativeSpent'] = cumulative_spent['CumulativeSpent'].values
df_clean['CumulativeOrders'] = cumulative_orders['CumulativeOrders'].values
df_clean['CumulativeSpent'] = df_clean['CumulativeSpent'].fillna(0)
df_clean['CumulativeOrders'] = df_clean['CumulativeOrders'].fillna(0)

cumulative_features = df_clean.groupby(['CustomerID', 'InvoiceYearMonth']).agg({
    'CumulativeSpent': 'first',
    'CumulativeOrders': 'first'
}).reset_index()

cumulative_features.columns = ['CustomerID', 'Month', 'CumulativeSpent', 'CumulativeOrders']
customer_monthly = customer_monthly.merge(cumulative_features, on=['CustomerID', 'Month'], how='left')
customer_monthly['CumulativeSpent'] = customer_monthly['CumulativeSpent'].fillna(0)
customer_monthly['CumulativeOrders'] = customer_monthly['CumulativeOrders'].fillna(0)

print(f"   - CumulativeSpent: общая сумма трат за всю историю ДО текущего месяца")
print(f"   - CumulativeOrders: общее количество заказов за всю историю ДО текущего месяца")

print("\n6. СКОЛЬЗЯЩИЕ СТАТИСТИКИ (за предыдущие 3 месяца):")

df_clean_sorted = df_clean.sort_values(['CustomerID', 'InvoiceDate'])
df_clean_sorted['TransactionAmount_prev'] = df_clean_sorted.groupby('CustomerID')['TransactionAmount'].shift(1)
df_clean_sorted['Prev3MonthsMean'] = df_clean_sorted.groupby('CustomerID')['TransactionAmount_prev'].transform(
    lambda x: x.rolling(90, min_periods=1).mean()
)
df_clean_sorted['Prev3MonthsStd'] = df_clean_sorted.groupby('CustomerID')['TransactionAmount_prev'].transform(
    lambda x: x.rolling(90, min_periods=1).std()
)
df_clean_sorted['Prev3MonthsStd'] = df_clean_sorted['Prev3MonthsStd'].fillna(0)

rolling_monthly = df_clean_sorted.groupby(['CustomerID', 'InvoiceYearMonth']).agg({
    'Prev3MonthsMean': 'first',
    'Prev3MonthsStd': 'first'
}).reset_index()

rolling_monthly.columns = ['CustomerID', 'Month', 'Prev3MonthsMean', 'Prev3MonthsStd']
customer_monthly = customer_monthly.merge(rolling_monthly, on=['CustomerID', 'Month'], how='left')
customer_monthly['Prev3MonthsMean'] = customer_monthly['Prev3MonthsMean'].fillna(0)
customer_monthly['Prev3MonthsStd'] = customer_monthly['Prev3MonthsStd'].fillna(0)

print(f"   - Prev3MonthsMean: средний чек за последние 3 месяца ДО текущего")
print(f"   - Prev3MonthsStd: изменчивость трат за последние 3 месяца")

print("\n7. ВРЕМЕННЫЕ ПРИЗНАКИ (не зависят от данных):")
df_clean['InvoiceDayOfWeek'] = df_clean['InvoiceDate'].dt.dayofweek
df_clean['InvoiceHour'] = df_clean['InvoiceDate'].dt.hour

preferred_day = df_clean.groupby('CustomerID')['InvoiceDayOfWeek'].agg(
    lambda x: x.mode().iloc[0] if len(x.mode()) > 0 else 0
).reset_index()
preferred_day.columns = ['CustomerID', 'PreferredDay']

preferred_hour = df_clean.groupby('CustomerID')['InvoiceHour'].agg(
    lambda x: x.mode().iloc[0] if len(x.mode()) > 0 else 0
).reset_index()
preferred_hour.columns = ['CustomerID', 'PreferredHour']

customer_monthly = customer_monthly.merge(preferred_day, on='CustomerID', how='left')
customer_monthly = customer_monthly.merge(preferred_hour, on='CustomerID', how='left')
customer_monthly['PreferredDay'] = customer_monthly['PreferredDay'].fillna(0)
customer_monthly['PreferredHour'] = customer_monthly['PreferredHour'].fillna(0)

country_feature = df_clean.groupby('CustomerID')['Country'].first().reset_index()
country_feature.columns = ['CustomerID', 'Country']
customer_monthly = customer_monthly.merge(country_feature, on='CustomerID', how='left')
customer_monthly['Country'] = customer_monthly['Country'].fillna('Unknown')

print(f"   - PreferredDay: любимый день недели")
print(f"   - PreferredHour: любимый час")
print(f"   - Country: страна клиента")

print("\n8. ДОПОЛНИТЕЛЬНЫЕ ВРЕМЕННЫЕ ПРИЗНАКИ:")
customer_monthly['Month_Num'] = customer_monthly['Month'].dt.month
customer_monthly['Quarter'] = customer_monthly['Month'].dt.quarter
customer_monthly['Year'] = customer_monthly['Month'].dt.year
customer_monthly['DaysInMonth'] = customer_monthly['Month'].dt.days_in_month
customer_monthly['IsDecember'] = (customer_monthly['Month_Num'] == 12).astype(int)
customer_monthly['IsWeekendPreferred'] = customer_monthly['PreferredDay'].isin([5, 6]).astype(int)

print(f"   - Month_Num, Quarter, Year, DaysInMonth, IsDecember, IsWeekendPreferred")

print("\n9. УДАЛЕНИЕ СТРОК С NaN (первые месяцы клиентов):")
before_drop = len(customer_monthly)
customer_monthly = customer_monthly.dropna(subset=['Spent_Lag_1', 'Spent_Lag_2', 'Spent_Lag_3'])
after_drop = len(customer_monthly)
print(f"   - Удалено: {before_drop - after_drop} строк")
print(f"   - Осталось: {after_drop} строк")

print("\n10. ПРОВЕРКА ДИАПАЗОНОВ ПРИЗНАКОВ:")
print(f"   - TotalSpent: {customer_monthly['TotalSpent'].min():.2f} to {customer_monthly['TotalSpent'].max():.2f}")
print(f"   - Spent_Lag_1: {customer_monthly['Spent_Lag_1'].min():.2f} to {customer_monthly['Spent_Lag_1'].max():.2f}")
print(f"   - CumulativeSpent: {customer_monthly['CumulativeSpent'].min():.2f} to {customer_monthly['CumulativeSpent'].max():.2f}")

final_df_clean = customer_monthly.copy()
print(f"\n✅ Итоговое количество признаков: {final_df_clean.shape[1]}")
print(f"✅ Итоговое количество строк: {final_df_clean.shape[0]:,}")
```

**One-hot encoding и разделение данных (train/val/test)**

Этот блок выполняет преобразование категориальных признаков и корректное разделение данных с избеганием data leakage

```python
print("=" * 60)
print("БЛОК 6: ONE-HOT ENCODING И РАЗДЕЛЕНИЕ ДАННЫХ")
print("=" * 60)

print("\n1. ПРОВЕРКА НАЛИЧИЯ КОЛОНОК:")
print(f"   - Колонки в final_df_clean: {final_df_clean.columns.tolist()}")
print(f"   - Количество строк: {len(final_df_clean):,}")

print("\n2. ПРЕОБРАЗОВАНИЕ КАТЕГОРИАЛЬНЫХ ПРИЗНАКОВ:")

if 'Country' in final_df_clean.columns:
    print(f"   - Найдена колонка Country, выполняем one-hot encoding")
    print(f"   - Уникальные страны: {final_df_clean['Country'].nunique()}")
    final_df_clean = pd.get_dummies(final_df_clean, columns=['Country'], prefix='Country', drop_first=True)
else:
    print(f"   - Колонка Country отсутствует (возможно уже закодирована)")

print(f"\n   - После преобразования: {final_df_clean.shape[1]} колонок")

print("\n3. ОПРЕДЕЛЕНИЕ ПРИЗНАКОВ И ЦЕЛЕВОЙ ПЕРЕМЕННОЙ:")
target_col = 'TotalSpent'

feature_cols = [col for col in final_df_clean.columns if col not in ['CustomerID', 'Month', target_col]]

print(f"   - Целевая переменная: {target_col}")
print(f"   - Количество признаков: {len(feature_cols)}")
print(f"   - Первые 10 признаков: {feature_cols[:10]}")

print("\n4. ПРОВЕРКА НА DATA LEAKAGE:")
print("   - Все признаки используют только данные из ПРОШЛОГО")
print("   - Нет признаков, заглядывающих в будущее")

print("\n5. РАЗДЕЛЕНИЕ ПО ВРЕМЕНИ:")
final_df_clean = final_df_clean.sort_values('Month')
print(f"   - Период данных: с {final_df_clean['Month'].min()} по {final_df_clean['Month'].max()}")

months_sorted = sorted(final_df_clean['Month'].unique())
train_months = months_sorted[:-4]
val_months = months_sorted[-4:-2]
test_months = months_sorted[-2:]

print(f"   - Обучающие месяцы: {train_months[0]} до {train_months[-1]} ({len(train_months)} месяцев)")
print(f"   - Валидационные месяцы: {val_months[0]} до {val_months[-1]} ({len(val_months)} месяцев)")
print(f"   - Тестовые месяцы: {test_months[0]} до {test_months[-1]} ({len(test_months)} месяцев)")

X = final_df_clean[feature_cols]
y = final_df_clean[target_col]

X_train = X[final_df_clean['Month'].isin(train_months)]
y_train = y[final_df_clean['Month'].isin(train_months)]

X_val = X[final_df_clean['Month'].isin(val_months)]
y_val = y[final_df_clean['Month'].isin(val_months)]

X_test = X[final_df_clean['Month'].isin(test_months)]
y_test = y[final_df_clean['Month'].isin(test_months)]

print(f"\n6. РАЗМЕРЫ ВЫБОРОК:")
print(f"   - Train: {X_train.shape[0]:,} строк ({X_train.shape[0]/len(X)*100:.1f}%)")
print(f"   - Val:   {X_val.shape[0]:,} строк ({X_val.shape[0]/len(X)*100:.1f}%)")
print(f"   - Test:  {X_test.shape[0]:,} строк ({X_test.shape[0]/len(X)*100:.1f}%)")
print(f"   - Всего: {len(X):,} строк")

print("\n7. ПРОВЕРКА РАСПРЕДЕЛЕНИЯ ЦЕЛЕВОЙ ПЕРЕМЕННОЙ ПО ВЫБОРКАМ:")
print(f"   - Train mean: £{y_train.mean():.2f}, median: £{y_train.median():.2f}")
print(f"   - Val mean:   £{y_val.mean():.2f}, median: £{y_val.median():.2f}")
print(f"   - Test mean:  £{y_test.mean():.2f}, median: £{y_test.median():.2f}")

print("\n8. МАСШТАБИРОВАНИЕ ПРИЗНАКОВ (fit только на train):")
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=feature_cols)
X_val_scaled = pd.DataFrame(X_val_scaled, columns=feature_cols)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=feature_cols)

print("   - StandardScaler обучен только на train данных")
print("   - Все признаки масштабированы (среднее=0, дисперсия=1)")

print("\n9. СОХРАНЕНИЕ ПОДГОТОВЛЕННЫХ ДАННЫХ:")
import pickle

processed_data = {
    'X_train': X_train_scaled,
    'X_val': X_val_scaled,
    'X_test': X_test_scaled,
    'y_train': y_train,
    'y_val': y_val,
    'y_test': y_test,
    'feature_names': feature_cols,
    'scaler': scaler,
    'train_months': train_months,
    'val_months': val_months,
    'test_months': test_months
}

with open('/content/processed_data.pkl', 'wb') as f:
    pickle.dump(processed_data, f)

print("   - Данные сохранены в '/content/processed_data.pkl'")

print("\n10. ФИНАЛЬНАЯ ПРОВЕРКА:")
print(f"    - X_train shape: {X_train_scaled.shape}")
print(f"    - X_val shape: {X_val_scaled.shape}")
print(f"    - X_test shape: {X_test_scaled.shape}")
print(f"    - Количество признаков: {len(feature_cols)}")

print("\n✅ БЛОК 6 ЗАВЕРШЕН. ДАННЫЕ ГОТОВЫ ДЛЯ МОДЕЛИРОВАНИЯ.")
```

**Выбор метрик качества и Baseline модель**

Этот блок определяет метрики для оценки регрессионных моделей и обучает Baseline (DummyRegressor) для сравнения.

 ВЫБРАННЫЕ МЕТРИКИ:
   
   а) MAE (Mean Absolute Error) - ГЛАВНАЯ МЕТРИКА

      - Интерпретируемость: "Ошибка в £"
      - Робастность к выбросам
   б) RMSE (Root Mean Squared Error)

      - Штрафует большие ошибки сильнее, чем MAE
      - Полезна для выявления моделей, которые иногда сильно ошибаются
   в) R² (Коэффициент детерминации)

      - Показывает, какую долю дисперсии объясняет модель
      - Удобен для сравнения с наивными моделями
      - Значение 0.7 означает, что модель объясняет 70% вариации трат
   г) MAPE (Mean Absolute Percentage Error) - дополнительно

      - Относительная ошибка в процентах
      - Понятна для презентаций: "Ошибка составляет X%"
      - Может быть проблематичен при нулевых значениях

```python
print("=" * 60)
print("БЛОК 7: ВЫБОР МЕТРИК КАЧЕСТВА И BASELINE МОДЕЛЬ")
print("=" * 60)

print("\n1. ОБОСНОВАНИЕ ВЫБОРА МЕТРИК:")

print("\n2. ЗАГРУЗКА ПОДГОТОВЛЕННЫХ ДАННЫХ:")
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

with open('/content/processed_data.pkl', 'rb') as f:
    data = pickle.load(f)

X_train = data['X_train']
X_val = data['X_val']
X_test = data['X_test']
y_train = data['y_train']
y_val = data['y_val']
y_test = data['y_test']
feature_names = data['feature_names']

print(f"Train: {X_train.shape[0]:,} samples")
print(f"Val:   {X_val.shape[0]:,} samples")
print(f"Test:  {X_test.shape[0]:,} samples")

print("\n3. ОПРЕДЕЛЕНИЕ ФУНКЦИЙ ДЛЯ ОЦЕНКИ МОДЕЛЕЙ:")

def evaluate_model(y_true, y_pred, model_name="Model"):
    mae = np.mean(np.abs(y_true - y_pred))
    mse = np.mean((y_true - y_pred) ** 2)
    rmse = np.sqrt(mse)
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    r2 = 1 - (ss_res / ss_tot) if ss_tot != 0 else 0

    y_true_nonzero = y_true[y_true != 0]
    y_pred_nonzero = y_pred[y_true != 0]
    if len(y_true_nonzero) > 0:
        mape = np.mean(np.abs((y_true_nonzero - y_pred_nonzero) / y_true_nonzero)) * 100
    else:
        mape = np.nan

    metrics = {'MAE': mae, 'RMSE': rmse, 'R2': r2, 'MAPE': mape}

    print(f"\n{'='*50}")
    print(f"{model_name}")
    print(f"{'='*50}")
    print(f"MAE:  £{mae:.2f}")
    print(f"RMSE: £{rmse:.2f}")
    print(f"R²:   {r2:.4f}")
    if not np.isnan(mape):
        print(f"MAPE: {mape:.2f}%")
    print(f"{'='*50}")

    return metrics

def plot_predictions(y_true, y_pred, model_name="Model"):
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))

    sample_idx = np.random.choice(len(y_true), min(500, len(y_true)), replace=False)

    axes[0].scatter(y_true.iloc[sample_idx], y_pred[sample_idx], alpha=0.5, s=10)
    min_val = min(y_true.min(), y_pred.min())
    max_val = max(y_true.max(), y_pred.max())
    axes[0].plot([min_val, max_val], [min_val, max_val], 'r--', alpha=0.7)
    axes[0].set_xlabel('Реальные значения (£)')
    axes[0].set_ylabel('Предсказанные значения (£)')
    axes[0].set_title(f'{model_name}\nРеальные vs Предсказанные')

    errors = y_pred - y_true
    axes[1].hist(errors, bins=50, color='steelblue', edgecolor='black', alpha=0.7)
    axes[1].axvline(0, color='red', linestyle='--', linewidth=2)
    axes[1].set_xlabel('Ошибка предсказания (£)')
    axes[1].set_ylabel('Частота')
    axes[1].set_title('Распределение ошибок')

    residuals = y_pred - y_true
    axes[2].scatter(y_true.iloc[sample_idx], residuals.iloc[sample_idx], alpha=0.5, s=10)
    axes[2].axhline(0, color='red', linestyle='--', linewidth=2)
    axes[2].set_xlabel('Предсказанные значения (£)')
    axes[2].set_ylabel('Остатки (£)')
    axes[2].set_title('График остатков')

    plt.tight_layout()
    plt.show()

    return fig

print("✅ Функции evaluate_model и plot_predictions созданы")

print("\n4. BASELINE МОДЕЛЬ (DummyRegressor):")
from sklearn.dummy import DummyRegressor

dummy_median = DummyRegressor(strategy='median')
dummy_median.fit(X_train, y_train)

y_pred_train_dummy = dummy_median.predict(X_train)
y_pred_val_dummy = dummy_median.predict(X_val)

print("\n📊 BASELINE (MEDIAN):")
train_metrics = evaluate_model(y_train, y_pred_train_dummy, "Train - Median Baseline")
val_metrics = evaluate_model(y_val, y_pred_val_dummy, "Validation - Median Baseline")

dummy_mean = DummyRegressor(strategy='mean')
dummy_mean.fit(X_train, y_train)
y_pred_val_mean = dummy_mean.predict(X_val)

print("\n📊 BASELINE (MEAN):")
val_metrics_mean = evaluate_model(y_val, y_pred_val_mean, "Validation - Mean Baseline")

baseline_mae = val_metrics['MAE']
print(f"\n✅ BASELINE MAE (медиана): £{baseline_mae:.2f}")
print("   Эта метрика будет использоваться для сравнения всех последующих моделей")

print("\n5. СОХРАНЕНИЕ BASELINE РЕЗУЛЬТАТОВ:")
baseline_results = {
    'model': dummy_median,
    'metrics': val_metrics,
    'predictions': y_pred_val_dummy
}

print(f"   - Baseline MAE сохранен: £{baseline_mae:.2f}")

plot_predictions(y_val, y_pred_val_dummy, "Baseline (Median)")
```

**Линейные модели (Ridge и Lasso)**

Этот блок обучает линейные модели с регуляризацией: Ridge (L2) и Lasso (L1)

```python
print("=" * 60)
print("БЛОК 8: ЛИНЕЙНЫЕ МОДЕЛИ (RIDGE И LASSO)")
print("=" * 60)

print("\n1. ЗАГРУЗКА ДАННЫХ:")
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, Lasso, RidgeCV, LassoCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

with open('/content/processed_data.pkl', 'rb') as f:
    data = pickle.load(f)

X_train = data['X_train']
X_val = data['X_val']
X_test = data['X_test']
y_train = data['y_train']
y_val = data['y_val']
y_test = data['y_test']
feature_names = data['feature_names']

print(f"Train: {X_train.shape[0]:,} samples, {X_train.shape[1]} features")
print(f"Val: {X_val.shape[0]:,} samples")
print(f"Test: {X_test.shape[0]:,} samples")

def evaluate_regression(y_true, y_pred, model_name="Model"):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)

    y_true_nonzero = y_true[y_true != 0]
    y_pred_nonzero = y_pred[y_true != 0]
    if len(y_true_nonzero) > 0:
        mape = np.mean(np.abs((y_true_nonzero - y_pred_nonzero) / y_true_nonzero)) * 100
    else:
        mape = np.nan

    print(f"\n{'='*50}")
    print(f"{model_name}")
    print(f"{'='*50}")
    print(f"MAE:  £{mae:.2f}")
    print(f"RMSE: £{rmse:.2f}")
    print(f"R²:   {r2:.4f}")
    if not np.isnan(mape):
        print(f"MAPE: {mape:.2f}%")

    return {'MAE': mae, 'RMSE': rmse, 'R2': r2, 'MAPE': mape}

print("\n2. МОДЕЛЬ 1: RIDGE РЕГРЕССИЯ (L2-регуляризация)")
print("   - Штрафует квадраты коэффициентов")
print("   - Хорошо работает при мультиколлинеарности")
print("   - Все признаки остаются в модели")

alphas = [0.1, 1, 10, 50, 100, 500, 1000]
ridge_cv = RidgeCV(alphas=alphas, scoring='neg_mean_absolute_error', cv=3)
ridge_cv.fit(X_train, y_train)

print(f"\n   Лучший alpha: {ridge_cv.alpha_}")

ridge_best = Ridge(alpha=ridge_cv.alpha_, random_state=42)
ridge_best.fit(X_train, y_train)

y_pred_train_ridge = ridge_best.predict(X_train)
y_pred_val_ridge = ridge_best.predict(X_val)
y_pred_test_ridge = ridge_best.predict(X_test)

ridge_train_metrics = evaluate_regression(y_train, y_pred_train_ridge, "Ridge - Train")
ridge_val_metrics = evaluate_regression(y_val, y_pred_val_ridge, "Ridge - Validation")
ridge_test_metrics = evaluate_regression(y_test, y_pred_test_ridge, "Ridge - Test")

ridge_coefs = pd.DataFrame({
    'feature': feature_names,
    'coefficient': ridge_best.coef_
}).sort_values('coefficient', ascending=False, key=abs)

print("\n   Топ-10 важных признаков (по модулю коэффициента):")
print(ridge_coefs.head(10).to_string(index=False))

print("\n3. МОДЕЛЬ 2: LASSO РЕГРЕССИЯ (L1-регуляризация)")
print("   - Штрафует абсолютные значения коэффициентов")
print("   - Обнуляет неважные признаки (отбор признаков)")
print("   - Хороша для разреженных данных")

lasso_cv = LassoCV(alphas=alphas, cv=3, random_state=42, max_iter=5000)
lasso_cv.fit(X_train, y_train)

print(f"\n   Лучший alpha: {lasso_cv.alpha_:.4f}")

lasso_best = Lasso(alpha=lasso_cv.alpha_, random_state=42, max_iter=5000)
lasso_best.fit(X_train, y_train)

y_pred_val_lasso = lasso_best.predict(X_val)
y_pred_test_lasso = lasso_best.predict(X_test)

lasso_val_metrics = evaluate_regression(y_val, y_pred_val_lasso, "Lasso - Validation")
lasso_test_metrics = evaluate_regression(y_test, y_pred_test_lasso, "Lasso - Test")

nonzero_features = np.sum(lasso_best.coef_ != 0)
print(f"\n   Количество признаков с ненулевыми коэффициентами: {nonzero_features} из {len(feature_names)}")

lasso_coefs = pd.DataFrame({
    'feature': feature_names,
    'coefficient': lasso_best.coef_
}).sort_values('coefficient', ascending=False, key=abs)
lasso_coefs = lasso_coefs[lasso_coefs['coefficient'] != 0]

print("\n   Ненулевые признаки (Lasso отобрал):")
print(lasso_coefs.to_string(index=False))

print("\n4. СРАВНЕНИЕ ЛИНЕЙНЫХ МОДЕЛЕЙ:")
comparison = pd.DataFrame({
    'Model': ['Ridge', 'Lasso'],
    'Validation MAE': [ridge_val_metrics['MAE'], lasso_val_metrics['MAE']],
    'Validation R²': [ridge_val_metrics['R2'], lasso_val_metrics['R2']],
    'Test MAE': [ridge_test_metrics['MAE'], lasso_test_metrics['MAE']],
    'Test R²': [ridge_test_metrics['R2'], lasso_test_metrics['R2']],
    'Nonzero Features': [len(feature_names), nonzero_features]
})
print(comparison.to_string(index=False))

print("\n5. ВИЗУАЛИЗАЦИЯ КОЭФФИЦИЕНТОВ:")
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

ridge_top15 = ridge_coefs.head(15)
axes[0].barh(range(len(ridge_top15)), ridge_top15['coefficient'].values, color='steelblue')
axes[0].set_yticks(range(len(ridge_top15)))
axes[0].set_yticklabels(ridge_top15['feature'].values)
axes[0].set_xlabel('Коэффициент')
axes[0].set_title('Ridge: Топ-15 важных признаков')

if len(lasso_coefs) > 0:
    lasso_top15 = lasso_coefs.head(min(15, len(lasso_coefs)))
    axes[1].barh(range(len(lasso_top15)), lasso_top15['coefficient'].values, color='coral')
    axes[1].set_yticks(range(len(lasso_top15)))
    axes[1].set_yticklabels(lasso_top15['feature'].values)
    axes[1].set_xlabel('Коэффициент')
    axes[1].set_title('Lasso: Важные признаки (ненулевые)')

plt.tight_layout()
plt.show()

print("\n6. ВИЗУАЛИЗАЦИЯ ПРЕДСКАЗАНИЙ (Lasso):")
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sample_idx = np.random.choice(len(y_val), min(300, len(y_val)), replace=False)
axes[0].scatter(y_val.iloc[sample_idx], y_pred_val_lasso[sample_idx], alpha=0.5, s=10)
min_val = min(y_val.min(), y_pred_val_lasso.min())
max_val = max(y_val.max(), y_pred_val_lasso.max())
axes[0].plot([min_val, max_val], [min_val, max_val], 'r--', alpha=0.7)
axes[0].set_xlabel('Реальные траты (£)')
axes[0].set_ylabel('Предсказанные траты (£)')
axes[0].set_title('Lasso: Реальные vs Предсказанные')

errors = y_pred_val_lasso - y_val
axes[1].hist(errors, bins=50, color='coral', edgecolor='black', alpha=0.7)
axes[1].axvline(0, color='red', linestyle='--', linewidth=2)
axes[1].set_xlabel('Ошибка предсказания (£)')
axes[1].set_ylabel('Частота')
axes[1].set_title('Lasso: Распределение ошибок')

plt.tight_layout()
plt.show()

print("\n7. СОХРАНЕНИЕ ЛУЧШЕЙ ЛИНЕЙНОЙ МОДЕЛИ:")
if ridge_val_metrics['MAE'] <= lasso_val_metrics['MAE']:
    best_linear = ridge_best
    best_linear_name = 'Ridge'
    best_linear_mae = ridge_val_metrics['MAE']
else:
    best_linear = lasso_best
    best_linear_name = 'Lasso'
    best_linear_mae = lasso_val_metrics['MAE']

with open('/content/best_linear_model.pkl', 'wb') as f:
    pickle.dump({
        'model': best_linear,
        'name': best_linear_name,
        'metrics': ridge_val_metrics if best_linear_name == 'Ridge' else lasso_val_metrics,
        'features': feature_names
    }, f)

print(f"   Лучшая линейная модель: {best_linear_name}")
print(f"   Validation MAE: £{best_linear_mae:.2f}")

print("\n✅ БЛОК 8 ЗАВЕРШЕН")
```

**Random Forest и XGBoost**

Этот блок обучает более сложные модели, которые должны показать лучшее качество, чем линейные модели.

```python
print("=" * 60)
print("БЛОК 9: RANDOM FOREST И XGBOOST")
print("=" * 60)

print("\n1. ЗАГРУЗКА ДАННЫХ:")
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

with open('/content/processed_data.pkl', 'rb') as f:
    data = pickle.load(f)

X_train = data['X_train']
X_val = data['X_val']
X_test = data['X_test']
y_train = data['y_train']
y_val = data['y_val']
y_test = data['y_test']
feature_names = data['feature_names']

print(f"Train: {X_train.shape[0]:,} samples, {X_train.shape[1]} features")
print(f"Val: {X_val.shape[0]:,} samples")
print(f"Test: {X_test.shape[0]:,} samples")

def evaluate_model(y_true, y_pred, model_name="Model"):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)

    y_true_nonzero = y_true[y_true != 0]
    y_pred_nonzero = y_pred[y_true != 0]
    if len(y_true_nonzero) > 0:
        mape = np.mean(np.abs((y_true_nonzero - y_pred_nonzero) / y_true_nonzero)) * 100
    else:
        mape = np.nan

    print(f"\n{'='*50}")
    print(f"{model_name}")
    print(f"{'='*50}")
    print(f"MAE:  £{mae:.2f}")
    print(f"RMSE: £{rmse:.2f}")
    print(f"R²:   {r2:.4f}")
    if not np.isnan(mape):
        print(f"MAPE: {mape:.2f}%")

    return {'MAE': mae, 'RMSE': rmse, 'R2': r2, 'MAPE': mape}

print("\n2. МОДЕЛЬ 1: RANDOM FOREST")
print("   - Ансамбль решающих деревьев")
print("   - Устойчива к переобучению")
print("   - Хорошо работает с табличными данными")

rf = RandomForestRegressor(
    n_estimators=100,
    max_depth=12,
    min_samples_split=20,
    min_samples_leaf=10,
    random_state=42,
    n_jobs=-1
)

print("\n   Обучение Random Forest...")
rf.fit(X_train, y_train)

y_pred_train_rf = rf.predict(X_train)
y_pred_val_rf = rf.predict(X_val)
y_pred_test_rf = rf.predict(X_test)

rf_train_metrics = evaluate_model(y_train, y_pred_train_rf, "Random Forest - Train")
rf_val_metrics = evaluate_model(y_val, y_pred_val_rf, "Random Forest - Validation")
rf_test_metrics = evaluate_model(y_test, y_pred_test_rf, "Random Forest - Test")

rf_importance = pd.DataFrame({
    'feature': feature_names,
    'importance': rf.feature_importances_
}).sort_values('importance', ascending=False)

print("\n   Топ-15 важных признаков (Random Forest):")
for i, row in rf_importance.head(15).iterrows():
    print(f"   {row['feature']}: {row['importance']:.4f}")

print("\n3. МОДЕЛЬ 2: XGBOOST")
print("   - Градиентный бустинг")
print("   - Обычно дает лучшее качество")
print("   - Требует больше настройки параметров")

xgb = XGBRegressor(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    verbosity=0,
    n_jobs=-1
)

print("\n   Обучение XGBoost...")
xgb.fit(X_train, y_train)

y_pred_train_xgb = xgb.predict(X_train)
y_pred_val_xgb = xgb.predict(X_val)
y_pred_test_xgb = xgb.predict(X_test)

xgb_train_metrics = evaluate_model(y_train, y_pred_train_xgb, "XGBoost - Train")
xgb_val_metrics = evaluate_model(y_val, y_pred_val_xgb, "XGBoost - Validation")
xgb_test_metrics = evaluate_model(y_test, y_pred_test_xgb, "XGBoost - Test")

xgb_importance = pd.DataFrame({
    'feature': feature_names,
    'importance': xgb.feature_importances_
}).sort_values('importance', ascending=False)

print("\n   Топ-15 важных признаков (XGBoost):")
for i, row in xgb_importance.head(15).iterrows():
    print(f"   {row['feature']}: {row['importance']:.4f}")

print("\n4. СРАВНЕНИЕ ВСЕХ МОДЕЛЕЙ:")
comparison = pd.DataFrame({
    'Model': ['Baseline (Median)', 'Lasso', 'Random Forest', 'XGBoost'],
    'Validation MAE': [816.76, 373.03, rf_val_metrics['MAE'], xgb_val_metrics['MAE']],
    'Validation R²': [-0.0307, 0.68, rf_val_metrics['R2'], xgb_val_metrics['R2']],
    'Test MAE': [None, None, rf_test_metrics['MAE'], xgb_test_metrics['MAE']],
    'Test R²': [None, None, rf_test_metrics['R2'], xgb_test_metrics['R2']]
})

print(comparison.to_string(index=False))

print("\n5. ВИЗУАЛИЗАЦИЯ ВАЖНОСТИ ПРИЗНАКОВ:")
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

rf_top15 = rf_importance.head(15)
axes[0].barh(range(len(rf_top15)), rf_top15['importance'].values, color='steelblue')
axes[0].set_yticks(range(len(rf_top15)))
axes[0].set_yticklabels(rf_top15['feature'].values)
axes[0].set_xlabel('Важность')
axes[0].set_title('Random Forest: Топ-15 важных признаков')
axes[0].invert_yaxis()

xgb_top15 = xgb_importance.head(15)
axes[1].barh(range(len(xgb_top15)), xgb_top15['importance'].values, color='coral')
axes[1].set_yticks(range(len(xgb_top15)))
axes[1].set_yticklabels(xgb_top15['feature'].values)
axes[1].set_xlabel('Важность')
axes[1].set_title('XGBoost: Топ-15 важных признаков')
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()

print("\n6. ВИЗУАЛИЗАЦИЯ ПРЕДСКАЗАНИЙ (XGBoost):")
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sample_idx = np.random.choice(len(y_val), min(300, len(y_val)), replace=False)
axes[0].scatter(y_val.iloc[sample_idx], y_pred_val_xgb[sample_idx], alpha=0.5, s=10)
min_val = min(y_val.min(), y_pred_val_xgb.min())
max_val = max(y_val.max(), y_pred_val_xgb.max())
axes[0].plot([min_val, max_val], [min_val, max_val], 'r--', alpha=0.7)
axes[0].set_xlabel('Реальные траты (£)')
axes[0].set_ylabel('Предсказанные траты (£)')
axes[0].set_title('XGBoost: Реальные vs Предсказанные')

errors = y_pred_val_xgb - y_val
axes[1].hist(errors, bins=50, color='coral', edgecolor='black', alpha=0.7)
axes[1].axvline(0, color='red', linestyle='--', linewidth=2)
axes[1].set_xlabel('Ошибка предсказания (£)')
axes[1].set_ylabel('Частота')
axes[1].set_title('XGBoost: Распределение ошибок')

plt.tight_layout()
plt.show()

print("\n7. СОХРАНЕНИЕ ЛУЧШЕЙ МОДЕЛИ:")
if xgb_val_metrics['MAE'] <= rf_val_metrics['MAE']:
    best_model = xgb
    best_model_name = 'XGBoost'
    best_model_mae = xgb_val_metrics['MAE']
else:
    best_model = rf
    best_model_name = 'Random Forest'
    best_model_mae = rf_val_metrics['MAE']

with open('/content/best_model.pkl', 'wb') as f:
    pickle.dump({
        'model': best_model,
        'name': best_model_name,
        'metrics': xgb_val_metrics if best_model_name == 'XGBoost' else rf_val_metrics,
        'features': feature_names
    }, f)

print("\n8. ФИНАЛЬНЫЕ ВЫВОДЫ:")

if best_model_name == 'XGBoost':
    best_r2 = xgb_val_metrics['R2']
else:
    best_r2 = rf_val_metrics['R2']

print(f"""
   - Baseline MAE: £816.76
   - Lasso MAE: £373.03 (улучшение на 54%)
   - Random Forest MAE: £{rf_val_metrics['MAE']:.2f}
   - XGBoost MAE: £{xgb_val_metrics['MAE']:.2f}

   ЛУЧШАЯ МОДЕЛЬ: {best_model_name}
   Ошибка предсказания: £{best_model_mae:.2f}
   Доля объясненной дисперсии (R²): {best_r2:.4f}
""")

print("\n✅ БЛОК 9 ЗАВЕРШЕН")
```
